```python
import seelib
import time
import numpy as np
a = seelib.random.pos_def(500)
np.set_printoptions(precision=2, linewidth= 200)
print(a)
t = time.time()
for i in range(100):
    U = seelib.linalg.cholesky(a)
print(time.time() - t)
t = time.time()
for i in range(100):
    U = seelib.linalg.jit.cholesky(a)
print(time.time() - t)
t = time.time()
for i in range(100):
    U = np.linalg.cholesky(a)
print(time.time() - t)
print(U @ U.T - a)
```

```python
import numpy as np
import seelib

np.set_printoptions(precision=2, linewidth= 200)
x = np.random.uniform(0,1,10)
x = np.sort(x)
y = np.cos(2 * np.pi * x)
X = np.linspace(0, 1, 500)

import matplotlib.pyplot as plt
import timeit
plt.plot(x, y, 'o')
Y = seelib.interp.cubic_spline(x, y, X, dm1=5, dm2=-5)
plt.plot(X, Y)
plt.show()
print(Y)

# timeit.timeit('seelib.interp.square_spline(x, y, X)', number= 10000, globals= globals())
```
